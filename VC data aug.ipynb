```python
t = text.lower()
t = t.encode('ascii', 'ignore').decode()
import re
def remove_other(x):
    x = re.sub("\$"," ", x) #remove $
    x = re.sub("https*\S+", " ", x) #remove url
    x = re.sub("\'\w+", '', x) #remove i'm,we're,let's after the '
    x = re.sub("[0-9]+", '', x) #remove numbers
    x = re.sub("[0-9]+", '', x) #remove numbers
    x = re.sub(r'[^\w\s]', '', x)
    return x

t=remove_other(t)
import nltk
from nltk.corpus import stopwords
#nltk.download("stopwords") #uncomment it when run it for the first time
stop_words = set(stopwords.words("english"))

t = " ".join([word for word in t.split() if word not in stop_words])
t
```

```python
import nltk
#nltk.download('wordnet') #uncommented for 1st time running
from nltk.stem import WordNetLemmatizer

words=t.split()

def lemmatize(words):
    lemmatizer = WordNetLemmatizer()
    nwords=[]
    for word in words:
        word=lemmatizer.lemmatize(word)
        nwords.append(word)
    return ' '.join(nwords)

t = lemmatize(words)
```

```python
t
```

```python
from sklearn.feature_extraction.text import CountVectorizer
t_list = [t]  # convert string to list

vec = CountVectorizer()
counts = vec.fit_transform(t_list)
counts = counts.toarray()
count_df = pd.DataFrame(counts, columns=vec.get_feature_names())
```

```python
count_df.head()
```

```python
import itertools
import collections

def count_words(text):
    all_words = list(itertools.chain(text.split()))
    counts = collections.Counter(all_words)
    counts_df = pd.DataFrame(counts.most_common(100),
                            columns=['words', 'count'])

    return counts_df
count_df = count_words(t)
count_df
```

```python
import plotly.express as px
fig = px.bar(count_df.head(10), x='words', y='count', title="Top 10 word frequency")
fig.show()
```

```python
# Libraries
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Create a list of word
text=' '.join(count_df.words)

# Create the wordcloud object
wordcloud = WordCloud(width=1500, height=1000, margin=0,background_color='white').generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```

```python
from nltk.util import ngrams
def count_2gram(text):
    all_words = list(itertools.chain(text.split()))
    esBigrams = ngrams(all_words, 2)
    counts = collections.Counter(esBigrams)
    counts_df = pd.DataFrame(counts.most_common(100),
                            columns=['words', 'count'])
    dictionary2 = [' '.join(tup) for tup in counts_df.words]
    counts_df.words=dictionary2

    return counts_df
counts_df2=count_2gram(t)
counts_df2.head(20)
```

```python
words_dict = dict(counts_df2.values)

wordCloud = WordCloud(max_words=200, height=1000, width=1500,background_color='white')
wordCloud.generate_from_frequencies(words_dict)
plt.title('Most frequently occurring bigrams connected by same colour and font size')
plt.imshow(wordCloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
